Now we can identify the contents of the data:
* User information
    * user id (studentId_clientType field)
    * city
    * country
    * university
    * course
    * state (core user, casual user, new user, cold user)
* Device information
    * device id (install_uuid field)
    * browser
    * new or old device
    * platform (windows, linux, ios)
    * operating system version
* Session information
    * session id (session_uuid field)
    * carrier operator
* Event information
    * event id (uuid field)
    * timestamp
    * name (page view)
    * page url
    * page categories
